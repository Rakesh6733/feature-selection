```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data for each bar graph
categories = ['A', 'B', 'C', 'D', 'E']
data1 = [5, 8, 6, 4, 7]
data2 = [3, 6, 4, 9, 2]
data3 = [2, 7, 5, 3, 6]

# Set the width of each bar
bar_width = 0.25

# Set the positions of the bars on the x-axis
bar_positions1 = np.arange(len(categories))
bar_positions2 = [x + bar_width for x in bar_positions1]
bar_positions3 = [x + 2 * bar_width for x in bar_positions1]

# Increase the width of the x-axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting the bar graphs
ax.bar(bar_positions1, data1, width=bar_width, label='Data 1')
ax.bar(bar_positions2, data2, width=bar_width, label='Data 2')
ax.bar(bar_positions3, data3, width=bar_width, label='Data 3')

# Adding labels, title, and legend
ax.set_xlabel('Categories')
ax.set_ylabel('Values')
ax.set_title('Multiple Bar Graphs')
ax.set_xticks(bar_positions2)
ax.set_xticklabels(categories)
ax.legend()

# Display the plot
plt.show()
```
