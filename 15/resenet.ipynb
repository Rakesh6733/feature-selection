```python
from tensorflow.keras.applications import ResNet50
resenet = ResNet50(input_shape =(967,469,7),weights=None,include_top=False)
```

```python

for layer in resenet.layers:
  layer.trainable = False
from tensorflow.keras.layers import Dense,Flatten
from tensorflow.keras.models import Model
     
```

```python
x = Flatten()(resenet.output)
predications = Dense(7,activation="softmax")(x)
model = Model(inputs=resenet.input,outputs=predications)

import tensorflow as tf

model.compile(optimizer=tf.keras.optimizers.Adam(0.001),
              loss='mse',
              metrics=['mae', 'mse'])
```
